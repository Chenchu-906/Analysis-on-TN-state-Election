# Benfords law

```python
data["leading_digit_votes"]=data["Votes"].apply(lambda X:str(X)[0])
```

```python
data['leading_digit_votes']=data['leading_digit_votes'].astype(int)
```

```python
import matplotlib.pyplot as plt


leading_digit_counts = data["leading_digit_votes"].value_counts()


(leading_digit_counts / leading_digit_counts.sum()).plot(kind='bar')

plt.xlabel("Leading Digit")
plt.ylabel("Proportion")
plt.title("Leading Digit Proportions by Benford's Law")
plt.show()
```

 Benford's Law can raise a flag for further investigation, it's crucial to remember that drawing conclusions solely based on this law can be misleading.

```python
import numpy as np
from scipy.stats import chisquare, zscore


expected_proportions = np.array([np.log10(1 + 1/d) for d in range(1, 10)])


total_count = leading_digit_counts.sum()
expected_frequencies = expected_proportions * total_count


observed_frequencies = leading_digit_counts.values

#  Chi-square test
chi2_stat, chi2_p_val = chisquare(observed_frequencies, expected_frequencies)

#  Z-test
observed_proportions = observed_frequencies / total_count
z_scores = (observed_proportions - expected_proportions) / np.sqrt(expected_proportions * (1 - expected_proportions) / total_count)


print("Chi-square test:")
print("Chi-square statistic:", chi2_stat)
print("P-value:", chi2_p_val)

print("\nZ-test:")
print("Z-scores for each leading digit:")
for i in range(9):
    print(f"Leading Digit {i+1}: {z_scores[i]}")
```
